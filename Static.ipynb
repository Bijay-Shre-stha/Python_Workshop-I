```python
tips=sns.load_dataset('tips')
tips.head(10)
```

```python
tips.info()
```

```python
plt.figure(figsize=(6,6))
sns.barplot(x='size' , y='tip' ,data=tips, hue='sex', errorbar=None, palette='summer')
plt.xlabel('Size of group')
plt.ylabel('Amount of Tips')
plt.title('Distribution of tips based on gender')
plt.show()
```

```python
# sns.scatterplot(x='tip', y='total_bill', hue='sex', data=tips)
# plt.ylabel('total_bill')
# plt.xlabel('tips')
sns.pairplot( data=tips , hue ='sex')
plt.show()
```

```python
sns.histplot(x='smoker', data=tips)
plt.show()
```

```python
sns.countplot(x='smoker', data=tips, palette =['red', 'green'])
plt.show()
```
